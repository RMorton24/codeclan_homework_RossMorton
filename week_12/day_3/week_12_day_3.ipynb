# Question 1
Can you join all 4 credit card datasets together so that youâ€™re left with a dataset that looks like below with 109 rows and 12 columns?

- CardBase.csv
- CustomerBase.csv
- FraudBase.csv
- TransactionBase.csv


```python
card_base.merge(
    customer_base, 
    how = "left", 
    on = "cust_id"
).merge(
    transaction_base, 
    how = "left", 
    left_on = "card_number", 
    right_on = "credit_card_id",
).merge(
    fraud_base,
    how = "right",
    on = "transaction_id"
).drop(columns = "credit_card_id")
```

# Question 2
Read in exam_scores and transform it into long format with two new columns exam_question and score. Remove superfluous information from the values in exam_question.
